```python
import requests
```

```python
BASE_URL = "http://127.0.0.1:8000"
```

# 1. POST: Añadir un nuevo chiste

```python
nuevo_chiste = {
        "categoria": "programadores",
        "texto": "¿Por qué los programadores confunden Halloween con Navidad? Porque Oct 31 == Dec 25."
    }
response = requests.post(f"{BASE_URL}/chiste/", json=nuevo_chiste)
print(f"Status: {response.status_code} | Respuesta: {response.json()}\n")
```

# 2. GET: Obtener un chiste aleatorio de la categoría creada

```python
print("2. Probando GET /chiste/programadores")
response = requests.get(f"{BASE_URL}/chiste/programadores")
if response.status_code == 200:
    print(f"Chiste recibido: {response.json()['chiste']}\n")
else:
    print(f"Error: {response.status_code}\n")
```

# 3. PUT: Reemplazar toda la categoría

```python
print("3. Probando PUT /chiste/programadores (Reemplazar lista)...")
lista_nueva = ["Chiste malo 1", "Chiste malo 2"]
response = requests.put(f"{BASE_URL}/chiste/programadores", json=lista_nueva)
print(f"Status: {response.status_code} | Respuesta: {response.json()}\n")
```

# 4. PATCH: Editar solo el primer chiste de esa lista (índice 0)

```python
print("4. Probando PATCH /chiste/programadores/0...")
nuevo_texto = "Este es un chiste de programadores mejorado."

# Pasamos el nuevo texto como parámetro de consulta (query param) según el script anterior
params = {"nuevo_texto": nuevo_texto}
response = requests.patch(f"{BASE_URL}/chiste/programadores/0", params=params)
print(f"Status: {response.status_code} | Respuesta: {response.json()}\n")
```

# 5. DELETE: Eliminar la categoría de prueba

```python
print("5. Probando DELETE /chiste/programadores...")
response = requests.delete(f"{BASE_URL}/chiste/programadores")
print(f"Status: {response.status_code} | Respuesta: {response.json()}\n")
```
